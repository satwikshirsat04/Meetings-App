```python
import librosa
import numpy as np
import librosa.display
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity
```

```python
y1, sr1 = librosa.load('audio.wav', sr=16000)
y2, sr2 = librosa.load('audio2.wav', sr=16000)

# Compute MFCCs
mfcc1 = librosa.feature.mfcc(y=y1, sr=sr1, n_mfcc=13)
mfcc2 = librosa.feature.mfcc(y=y2, sr=sr2, n_mfcc=13)

# Take mean over time axis
mfcc1_mean = np.mean(mfcc1, axis=1)
mfcc2_mean = np.mean(mfcc2, axis=1)
```

```python
# ---- PLOT MFCC HEATMAPS ----
plt.figure(figsize=(10, 6))

# Audio 1 MFCC
plt.subplot(2, 1, 1)
librosa.display.specshow(mfcc1, x_axis='time', sr=sr1)
plt.colorbar(format='%+2.0f dB')
plt.title('MFCCs of Audio 1')
plt.xlabel('Time')
plt.ylabel('MFCC Coefficients')

# Audio 2 MFCC
plt.subplot(2, 1, 2)
librosa.display.specshow(mfcc2, x_axis='time', sr=sr2)
plt.colorbar(format='%+2.0f dB')
plt.title('MFCCs of Audio 2')
plt.xlabel('Time')
plt.ylabel('MFCC Coefficients')

plt.tight_layout()
plt.show()
```

```python
# Compute and print cosine similarity between mean MFCC vectors
similarity = cosine_similarity([mfcc1_mean], [mfcc2_mean])
print("Similarity:", float(similarity[0][0]))
```
